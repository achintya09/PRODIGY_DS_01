###IMPORTING LIBRARIES

###REMOVING COLUMNS WHICH ARE NOT USEFUL

```python
# prompt: drop Indicator Name and Indicator code from the dataset

dataset.drop(["Indicator Name", "Indicator Code", "Country Code"], axis=1, inplace=True)
```

###VISUALISING THE POPULATION CHANGE OVER THE YEARS OF A COUNTRY WITH LINE CHART

```python
country_name = input("enter the name of the country you want to show (e.g India):   ")

# Extract data for the specified country
country_data = dataset[dataset['Country Name'] == country_name]

# Extract years and population values
years = country_data.columns[1:]  # Assuming the years start from the second column
population = country_data.values.flatten()[1:]

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(years, population, marker='o', linestyle='-')
plt.title(f'Population of {country_name} (1960-2022)')
plt.xlabel('Year')
plt.ylabel('Population')
#plt.grid(True)
plt.show()
```

### VISUALISING THE TOP 10 COUNTRIES WHICH HAVE LEAST POPULATION

```python
latest_year = '2022'

# Sort the DataFrame based on the population of the latest year
dataset_sorted = dataset.sort_values(by=latest_year, ascending=True)

# Select the top 10 countries
top_10_countries = dataset_sorted.head(10)

# Plotting the bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_10_countries['Country Name'], top_10_countries[latest_year])
plt.title(f'Top 10 Countries with Lowest Population ({latest_year})')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45, ha='right')  # Rotate country names for better visibility
plt.show()
```

### PREDICT ANY COUNTRIES POPULATION OF 2022 AND MATCH IT WITH THE  ORIGINAL POPULATION OF THE SAME YEAR

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor


country_to_predict = input("enter the name of the country you want to predict(e.g India):   ")

# Extract data for the specified country
country_data = dataset[dataset['Country Name'] == country_to_predict].squeeze()

# Check if the dataset is empty for the specified country
if country_data.empty:
    print(f"No data found for {country_to_predict}. Check the country name.")
else:
    # Assuming the years start from the second column
    years = country_data.index[1:62].astype(int).to_numpy()  # Convert to NumPy array
    populations = country_data.values[1:62]

    # Check if there is enough data to train the model
    if len(years) < 2:
        print("Insufficient data to train the model.")
    else:
        # Prepare the data for training
        X = years.reshape(-1, 1)  # Reshape to make it a 2D array
        y = populations

        # Train a decision tree regression model
        model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
        model.fit(X, y)

        # Predict the population for 2022
        population_2022_predicted = model.predict([[2022]])

        # Original population of 2022
        original_population = country_data.values[-1]

        # Print the predicted population
        print(f"Predicted population of {country_to_predict} in 2022: {population_2022_predicted[0]:.0f}")
        print(f"Original  population of {country_to_predict} in 2022: {original_population} ")

        error = (original_population - round(population_2022_predicted[0]))
        accuracy = round((1 - (error/original_population))*100, 2)
        print(f"accuracy is:   {accuracy} %")
```

### PREDICTING THE POPULATION OF A PARTICULAR COUNTRY FOR THE YEAR 2023 AND SHOWING THE PERCENTAGE CHANGE WITH RESPECT TO PREVIOUS YEARS.

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor

country_to_predict = input("enter the name of the country you want to predict(e.g India):   ")

# Extract data for the specified country
country_data = dataset[dataset['Country Name'] == country_to_predict].squeeze()

# Check if the dataset is empty for the specified country
if country_data.empty:
    print(f"No data found for {country_to_predict}. Check the country name.")
else:
    # Assuming the years start from the second column
    years = country_data.index[1:].astype(int).to_numpy()  # Convert to NumPy array
    populations = country_data.values[1:]

    # Check if there is enough data to train the model
    if len(years) < 2:
        print("Insufficient data to train the model.")
    else:
        # Prepare the data for training
        X = years.reshape(-1, 1)  # Reshape to make it a 2D array
        y = populations

        # Train a decision tree regression model
        model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
        model.fit(X, y)

        # Create a new data point for the year 2023
        X_2023 = np.array([[2023]])

        # Predict the population for 2023
        population_2023_predicted = model.predict(X_2023)

        pred_2023 = round(population_2023_predicted[0])

        # Print the predicted population
        print(f"Predicted population for 2023: {pred_2023}")


        user_year = int(input("enter the year you want comparison with: "))
        pop_year = country_data.values[(user_year-2023)]
        change = pred_2023 - pop_year

        percent_change = round((change/pop_year)*100, 4)
        if percent_change < 0:
          print(f"percentage decrease in population: {-(percent_change)} % ")
        else:
          print(f"percentage increase in population: {percent_change} % ")
```
